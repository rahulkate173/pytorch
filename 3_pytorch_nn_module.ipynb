## nn module - pytorch
### nn module in pytorch provide us builtin components to develop neural network (like keras)

```python
import torch
import numpy
```

```python
## nn module
import torch.nn as nn
```

### creating and perceptron
image.png

```python
class model_(nn.Module): #basemodel = nn.module
  def __init__(self,num_feature):
    super().__init__()
    self.linear = nn.Linear(num_feature,1) ## input_feature,output_feature
  def forward(self,X):
    return self.linear(X)
```

```python
# constants
learning_rate = 0.001
epochs = 50
# tensor
X = torch.rand(5,5,dtype=torch.float32)
y = torch.tensor([1,1,0,0,1],dtype=torch.float32).reshape(5,1)
model = model_(X.shape[0])
model(X)
print(f"weights:\n{model.linear.weight}")
print(f"bias:\n{model.linear.bias}")
```

```python
## builtin loss
loss = nn.BCELoss() # binary cross entropy loss
```

```python
## built in optimizer
optimizer = torch.optim.SGD(
    model.parameters(), # the weights and bais that model will require
    learning_rate,
)
```

```python
for epoch in range(epochs):
  y_pred = model(X) #predcition
  Loss = loss(y_pred,y) # bactch loss
  Loss.backward() # gradient compute
  #print(model.linear.weight.grad.norm())
  optimizer.step() # gradient update
  optimizer.zero_grad() # clear previous gradient
  print(f"Epochs:{epoch+1},loss:{Loss}")
```
