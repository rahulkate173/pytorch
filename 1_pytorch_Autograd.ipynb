# AutoGrad
## Atomatic differentation option while defining tensor gives us ablity to get manula backtracking and built neural network from sratch

```python
import torch
```

```python
## torch.tensor(requires_grad=True) such that we can perform backtracking
a = torch.tensor(10,requires_grad=True,dtype=torch.float32)
a
```

```python
## y = x^2
## y' = 2*x
## example : x = 6 , y' = 12
x = torch.tensor(6,requires_grad=True,dtype=torch.float32)
y = x**2
y.backward() #   to calculate the gradient
x.grad # to set the value inside and get the answer
```

```python
## Perceptron
## y = x*w + b
## x = the input tensor , w = weights , b = bias

x = torch.rand(1,dtype=torch.float32)
w = torch.rand(1,requires_grad=True,dtype=torch.float32)
b = torch.rand(1,requires_grad=True,dtype=torch.float32)
z = x*w + b
y = torch.sigmoid(z)
# computing gradient
y.backward() # automatically compute
print(w.grad) # gradient calculation
print(b.grad)
print(x)
```
