```python
import torch
```

## Introduction to tensor

```python
## tensor = multidimensional array similar to numpy
### Built in tensor
print("Empty tensor\n",torch.empty(2,3))
print("Zeros tensor\n",torch.zeros(2,3))
print("random tensor\n",torch.rand(2,3))
print("Ones tensor\n",torch.ones(2,3))
print("identity matrix",torch.eye(2,3)) # n=2,m=3
seed = torch.manual_seed(1000)
print("to get reproducibility:",)
print("random tensor 1 \n",torch.rand(2,3))
print("random tensor 2 \n",torch.rand(2,3))
```

```python
# dtype
a = torch.tensor(1)
type(a)
```

```python
## custom tensor
a = torch.tensor([1,2,3])
print("Custom tensor:\n",a)
# 1. zero dimensional tensor
zero = torch.tensor(1)
print("Dimension:",zero.dim())
# 2. One dimension tensor
one = torch.tensor([1,2,3])
print("Dimension:",one.dim())
# 3. two dimension tensor
two = torch.tensor([[1,2,3],[1,2,3]])
print("Dimension:",two.dim())
# 4. three dimension tensor
three = torch.tensor([[[1,2,3],[1,2,3],[1,2,3]]])
print("Dimension:",three.dim())
## so on
```

```python
## Changing the tensor datatype
a = torch.tensor([1,2,3,4])
print("type",a.dtype)
a = a.to(torch.float32)
print("type",a.dtype)
```

```python
## Coping the tensor
a = torch.tensor([1,2,3])
b = a
b[0] = 4
print(a,b) ## the changes are reflected
```

```python
a = torch.tensor([1,2,3])
b = a.clone()
b[0] = 4
print(a,b) ## not changing
```

```python
## arithematic operation on tensor with constant
a = torch.tensor([[1,2,3],[4,5,6]])
print("location wise addition:",2+a)
print("scaling",2*a)
print("dividing",a/4)
```

```python
## arithematic operation on tensor with other tensor
a = torch.tensor([[1,2,3],[4,5,6]])
b = torch.tensor([[1,2,3],[4,5,6]])
print("Addition:",a+b)
print("Substraction:",a-b)
print("Multplication:",a*b)
print("Division",a/b)
print("Power operation",a**b)
```

```python
## tensor operation to get GPU
# to check cuda
print(torch.cuda.is_available())
# to set gpu
device = torch.device("cuda")
print(device)
# Creating on GPU
torch.tensor([1,2,3],device=device)
## moving cpu to gpu
b = torch.tensor([1,2,3])
a = b.to(device)
```

```python
## Reshaping the tensor
### Note:= while reshaping just get the multpilcation of new tensor(all shape) is equal to previous tensor(all shape)
a = torch.rand(4,4) # 4x4 = 16
print("Tensor\n:",a)
print("Shape:",a.shape)
print("reshaped tensor:\n",a.reshape(2,2,2,2)) # 2x2x2x2 = 16
print("Shape:",a.reshape(2,2,2,2).shape)
```

```python
## flatten the tensor
torch.rand(3,3).flatten()
```

```python
## Unsqeeze = adding one dimensional to selected position
### application = adding the batch while image processing and testing in CNN
image = torch.rand(256,256,3)
print("Image shape:",image.shape)
print("at zero positon",torch.unsqueeze(image,0).shape)
print("at one positon",torch.unsqueeze(image,1).shape)
```

```python
## Squeeze to remove any dimension
image = torch.rand(1,20)
print("Image shape:",image.shape)
torch.squeeze(image,0).shape
```

```python
## permute : Some permutation on the tensor
a = torch.rand(2,3,4)
## 0 | 1 | 2 = index of shape
## 2 | 3 | 4 = shape id of
b = a.permute(1,2,0) # shape = (3,4,2)
print(b.shape)
```
